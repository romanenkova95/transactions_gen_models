```python
import pandas as pd

from data_preprocessing.utils import preprocess_features, preprocess_targets, merge, save_parquet
```

```python
path_to_data = "../data/age/transactions_train.csv"
data = pd.read_csv(path_to_data, sep=",")
data.head()
```

```python
features_age = preprocess_features(
    data=data,
    mcc="small_group",
    client_id="client_id",
    date="trans_date",
    amount="amount_rur",
    prepare_local_targets=False
)
features_age
```

```python
path_to_targets = "../data/age/train_target.csv"
targets = pd.read_csv(path_to_targets, sep=",")

targets_age = preprocess_targets(data=targets, client_id="client_id", target="bins")
targets_age
```

```python
preprocessed_age = merge(features_age, targets_age)
preprocessed_age.head()
```

```python
save_parquet(res_df=preprocessed_age, path_to_folder="../data/local_targets", dataset_name="age")
```
